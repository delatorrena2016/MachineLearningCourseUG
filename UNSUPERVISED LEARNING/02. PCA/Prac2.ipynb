```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import mpl_toolkits.mplot3d
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
X = iris.data[:, :4]  # 150 puntos de dim 4
#Xprom = np.mean(X,2)
y = iris.target
n=np.size(X,1) #eje 1 tama√±o de dim, eje 0 no. puntos 
```

```python
pca=PCA(n_components=4)
X_4D=pca.fit_transform(X)

pca=PCA(n_components=3)
X_3D=pca.fit_transform(X)


pca=PCA(n_components=2)
X_2D=pca.fit_transform(X)
```

```python
kmeans_pca = KMeans(n_clusters = 3)
X4_clustered = kmeans_pca.fit_predict(X_4D)
X3_clustered = kmeans_pca.fit_predict(X_3D)
X2_clustered = kmeans_pca.fit_predict(X_2D)

LABEL_COLOR_MAP = {0 : 'r',1 : 'g',2 : 'b'}
label_color = [LABEL_COLOR_MAP[l] for l in X4_clustered]


plt.figure(figsize = (7,7))
plt.scatter(X_4D[:,0],X_4D[:,2], c= label_color, alpha=0.5) 
plt.show()
```

```python
LABEL_COLOR_MAP = {0 : 'r',1 : 'g',2 : 'b'}
label_color = [LABEL_COLOR_MAP[l] for l in X3_clustered]


plt.figure(figsize = (7,7))
plt.scatter(X_3D[:,0],X_3D[:,2], c= label_color, alpha=0.5) 
plt.show()
```

```python
LABEL_COLOR_MAP = {0 : 'r',1 : 'g',2 : 'b'}
label_color = [LABEL_COLOR_MAP[l] for l in X2_clustered]


plt.figure(figsize = (7,7))
plt.scatter(X_2D[:,0],X_2D[:,1], c= label_color, alpha=0.5) 
plt.show()
```
